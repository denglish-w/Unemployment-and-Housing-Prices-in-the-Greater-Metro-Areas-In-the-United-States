```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
from src.cleaning import DataCleaner
```

```python
# import the csv files
raw_unemployment_data = pd.read_csv('data/raw/ssamatab1.csv', skiprows=2).dropna()
housing_value_data = pd.read_csv('data/raw/hpi_po_metro.csv')
```

```python
data_cleaner = DataCleaner('data/raw/hpi_po_metro.csv', 'data/raw/ssamatab1.csv')
data_cleaner.rename_unemployment_columns(raw_unemployment_data)
data_cleaner.clean_unemployment_data
```

```python
# create data cleaner
unemployment_cleaner = DataCleaner('data/raw/ssamatab1.csv')
```

```python
# rename, drop null values, and type cast columns in the unemployment dataset

clean_unemployment_data = (
                            unemployment_cleaner
                           .rename_unemployment_columns()
                           .drop_n_values()
                           .unemployment_object_columns_to_int64()
                           .unemployment_float_cols_to_int64()
                           )
clean_unemployment_df = clean_unemployment_data.df
clean_unemployment_data
```

```python
# make SQL database with tables "Unemployment" and "Housing_value"

with sqlite3.connect("database.db") as conn:
    clean_unemployment_df.to_sql("Unemployment", conn, if_exists='replace', index=False)
    housing_value_data.to_sql("Housing_value", conn, if_exists='replace', index=False)
```

```python
query1 = """
SELECT * FROM Unemployment
"""
result1 = pd.read_sql(query1, conn)
result1
```
